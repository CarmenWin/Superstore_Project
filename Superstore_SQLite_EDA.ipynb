```python
import sqlite3
import pandas as pd

# Establish a connection to the SQLite database
con = sqlite3.connect("products.db")

# Execute the SQL query to calculate total sales by category
query = """
SELECT Category, SUM(Sales) AS TotalSales
FROM csv_superstore
GROUP BY Category
ORDER BY TotalSales DESC
LIMIT 1
"""

# Fetch the result of the query into a DataFrame
result = pd.read_sql(query, con)

# Extract the category with the maximum sales revenue
category_max_sales = result["Category"].iloc[0]

# Display the result
print("Category with Maximum Sales Revenue:", category_max_sales)
```

```python
#Removing outliers from dataset 

# Load the dataset from CSV
df = pd.read_csv('/Users/CW/Desktop/0_CAB/Superstore_Project/Superstore_data.csv', encoding='unicode_escape')


# Identify the numerical columns for outlier analysis
numeric_columns = ['Postal Code', 'Sales', 'Quantity', 'Discount']  

# Calculate z-scores for each value in the numeric columns
z_scores = stats.zscore(df[numeric_columns])

# Define the threshold for outliers (e.g., 3 standard deviations)
threshold = 3

# Filter the DataFrame to exclude rows with outliers
df_no_outliers = df[(abs(z_scores) < threshold).all(axis=1)]

# Optional: Create a new DataFrame with outliers removed
# df_no_outliers = df[(abs(z_scores) < threshold).all(axis=1)].copy()

# Optional: Overwrite the original DataFrame with outliers removed
# df = df[(abs(z_scores) < threshold).all(axis=1)]

# Print the summary or perform further analysis on the dataset without outliers
print(df_no_outliers.describe())
```
